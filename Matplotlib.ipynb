```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
axes.set_xlabel('X label')
axes.set_ylabel('y label')
axes.set_title('Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title('LARGER PLOT')

axes2.plot(y,x)
axes2.set_title('SMALLER PLOT')
```

```python
fig,axes= plt.subplots(nrows=1,ncols=2)

for current_ax in axes:
    current_ax.plot(x,y)
#axes.plot(x,y)
```

```python
fig,axes= plt.subplots(nrows=1,ncols=2) #tuple unpacking

axes[0].plot(x,y)
axes[0].set_title('First plot')
axes[1].plot(y,x)
axes[1].set_title('2nd plot')

plt.tight_layout()
```

```python
#2 step process
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='square')
ax.plot(x,x**3,label = 'cube')

ax.legend() #can specify loc see cell below, site has param to check
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='square')
ax.plot(x,x**3,label = 'cube')

ax.legend(loc=0) 
```

```python
fig = plt.figure()
ax= fig.add_axes([0,0,1,1])

ax.plot(x,y,color='g') #can take common color strings like blue organge etc
```

```python
#to get more detailed can put rgb hex codes
```

```python
fig = plt.figure()
ax= fig.add_axes([0,0,1,1])

ax.plot(x,y,color='#FF8C00',linewidth=2, alpha=0.5) #width to get thicker lines
```

alpha to control line transparency 

```python
fig = plt.figure()
ax= fig.add_axes([0,0,1,1])

ax.plot(x,y,color='#FF8C00',linewidth=2,linestyle='--')
```

```python
fig = plt.figure()
ax= fig.add_axes([0,0,1,1])

ax.plot(x,y,color='#FF8C00',linewidth=2, ls = '-', marker='*',markersize = 5
       ,markerfacecolor='yellow',markeredgewidth=3,markeredgecolor='green')
```
