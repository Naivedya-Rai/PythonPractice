```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.set_style('whitegrid')
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
#distribution joint plot showing 2 distri on top of each other
```

```python
sns.jointplot(x='fare',y='age',data=titanic)
```

```python
#distribution dist plot
```

```python

import warnings
warnings.filterwarnings('ignore')
```

```python
sns.distplot(titanic['fare'])
```

```python
#have to remove kde
```

```python
sns.distplot(titanic['fare'],kde=False,color='red',bins=30)
```

```python
#box plt class on x axis w age on y
```

```python
sns.boxplot(x='class',y='age',data=titanic,palette='rainbow')
```

```python
#swarm plot
```

```python
sns.swarmplot(x='class',y='age',data=titanic)
```

```python
#bar plot but it is a count plot w count y and sex x
```

```python
sns.countplot(x='sex',data=titanic)
```

```python
#corr plot and heat map
```

```python
sns.heatmap(titanic.corr(),cmap='coolwarm')
plt.title('titanic.corr()')
```

```python
#facet grid w two columns
```

```python
g=sns.FacetGrid(data=titanic,col='sex')
g.map(plt.hist,'age')
```

```python
g=sns.FacetGrid(data=titanic,col='sex')
g.map(sns.distplot,'age')
```

```python
g=sns.FacetGrid(data=titanic,col='sex')
g.map(sns.distplot,'age')
```
